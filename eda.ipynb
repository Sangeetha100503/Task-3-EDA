```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.hist(figsize=(10,6))
plt.show()
```

```python
sns.countplot(x='Species', data=df)
plt.show()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

### Summary
- Dataset is clean with no missing values
- Petal features are highly correlated with species
- Petal length and width are important for prediction
- Visualization helped identify feature relationships
